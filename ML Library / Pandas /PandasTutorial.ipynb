### Data Cleaning with Pandas
bad data can be 
- empty cells
- data in wrong format
- wrong data
- Duplicates 

````python
#Lets process some data available  at */data_preprocess_pandas.gsheet
````

````python
import pandas as pandas
````

````python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
df
````

problems with our data are
- data set contains some empty cells ("Date" in row 22 and calories in row 18 and 28
- the data set contains worng format ("date in row 26)
- the data set contains wrong data ("duratoin in row7)
- the data set contains duplicates (row 11 and 12)

````python
#return a DataFrame with no empty cells:
# df.dropna() will remove empty cells but will not change the dataframe, if we want data frame to be changed, use inplace=true
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
new_df=df.dropna()
new_df
````

````python
#remove the data having empty cells; it will remove the entire row
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
df.dropna(inplace=True)
print(df)
````

````python
#another way of handling the null values is by replacing
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
df.fillna(130,inplace=  True)
````

````python
# if we want to replace specific column only
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
df["Calories"].fillna(130, inplace=True)
print(df)
````

````python
#if we want to replace null vallue by mean, median or mode
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")

#using mean first in column Calories
x=df["Calories"].mean()
#the mean value is found to be x= 304.68

df["Calories"].fillna(x,inplace=True)
df
````

````python
#what if we want to replace null value by the median
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
med=df["Calories"].median()
df["Calories"].fillna(med,inplace=True)
df
````

Handle wrong data format:
- either remove the rows or convert all the cells in the columns into the same format.
- in our data, row 22 and row 26 the date has wrong format

````python
#lets try date column in the same format using to_datetime() method
````

````python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandasData.csv")
df["Date"]=pd.to_datetime(df["Date"])
print(df)
````

we can see still row 22 has NAt so we can simply remove it using dropna(). make sure following syntax is used for removing the Nan

````python
df.dropna(subset=["Date"],inplace=True)
df
````

Fixing wrong value or deal with the outliers:
- we can replace them with something else
- so in our case we will replace 450 by simply 45 in 7th row

for the small dataset we can directly set the value as

````python
df.loc[7,'Duration']=45
df
````

what if we have multiple data and the dataset is larger?
- loop through the entire columns
- eg, we can replace this value if greater than 60 by 60


````python
for x in df.index:
  if df.loc[x,"Duration"]>60:
    df.loc[x,"Duration"]=60
````

````python
df
````

````python
#next way is to delete the row
for x in df.index:
  if df.loc[x,"Duration"]>60:
    df.drop(x,inplace=True)
````

Handles the duplicates values in dataframe
- in our dataframe, row 11 and 12 are same, so we can remove the duplicate using ```dupllicated()``` method

````python
print(df.duplicated())
````

````python
#to remove duplicates use drop_duplicates() method
df.drop_duplicates(inplace=True)
df
````
